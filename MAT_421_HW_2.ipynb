# 17.1: Problem Statement
The main concept behind interpolation is that we need an interpolation function, y_hat, that can estimate accurately the data points within a data set such that: y_hat(xi)=yi, where yi is the dependent variable data values and xi being the independent data values. 

Python Code Example: Here we will use python to plot the interpolation of a data set at x = 8.5, given the data sets x=[5,7,10,13] and y=[10,13,16,19]

```python
import numpy as np
import matplotlib.pyplot as plt

x=[5,7,10,13]
y=[10,13,16,19]

x_interpolated = 8.5
y_hat = np.interp(x_interpolated, x, y)
print(y_hat)
```

# 17.2: Linear Interpolation: 
In this method the estimaed point is assumed to lie on the lie joining the nearest points to the left and right. 

- Assumptions Needed: For this method to work we need to make the crucial assumption that: xi<xi+1 and that xi<x<xi+1

The interpolation function for linear interpolation is as follows:

y(x) = y_i + ( (y_iplus1 - y_1)*(x - x_i) ) / (x_iplus1 - x_i)

Example: We can find the linear interpolation at x = 10 , in the data sets x = [5,7,9,11,13] and y = [10,11.5,15, 19 ,16] using python's scipy library using the fucntion interp1d

```python
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')

x = [5, 7, 9, 11, 13]
y = [10, 11.5, 15, 19, 16]
f = interp1d(x,y)
y_linearInterp = f(10)
print(y_linearInterp)

plt.figure(figsize = (10,8))
plt.plot(x, y, '-ob', 10, y_linearInterp, 'ro')
plt.title('Linear INterpolation at x = 10')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

# 17.3 Cubic Spline Interpolation
Cubic Spline: Interpolating function here is a set of piecewise cubic fucntions. For this to work we assume that (x_i,y_i) and (x_i+1, y_i+1) are joined together by the following polynomial:

S_i(x) = a_i* x^3 + b_i* x^2 + c_i*x + d_i

which is valid for x_i<=x<=x_i+1 for =1,..,n-1



Solving: Involves numerous steps but can break it down into the following steps: 


*   Determine the coefficeints a_i, b_i, c_i, and d_i for each cubic function
*   Given n points, there are a total of n-1 cubic fucntions to connect all the points needed 
*    So we need to find 4(n-1) unknowns and 4(n-1) independent equations we find the coefficients from



Example: Find the cubic Interpolation for the data sets, x = [1, 3, 5, 7, 9] and y =[10, 4, 8, 3, 7] for 0<=x<=10

```python
from scipy.interpolate import CubicSpline
import numpy as np 
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')

x = [1, 3, 5, 7, 9]
y = [10, 4, 8, 3, 7]

f = CubicSpline(x,y, bc_type = 'natural')
x_new = np.linspace(0, 10, 100)
y_new = f(x_new)

plt.figure(figsize = (10, 8))
plt.plot(x_new, y_new, 'b')
plt.plot(x, y, 'ro')
plt.title('Cubic Spline Interpolation')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
